```python
class ListNode(object):
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
# merge two reverted sorted linked lists

def merge(head1, head2):
    fake_head = ListNode(None)
    cur = fake_head
    while head1 and head2:
        if head1.value >= head2.value:
            cur.next = head1
            head1 = head1.next
        else:
            cur.next = head2
            head2 = head2.next
        cur = cur.next
    if head1:
        cur.next = head1
    if head2:
        cur.next = head2
    return fake_head.next
```

```python
# find the middle node in a singly linked list
def find_middle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow
```

```python
# merge sort linked list
def merge_sort(head):
    def merge(head1, head2):
        fake_head = ListNode(None)
        cur = fake_head
        while head1 and head2:
            if head1.value >= head2.value:
                cur.next = head1
                head1 = head1.next
            else:
                cur.next = head2
                head2 = head2.next
            cur = cur.next
        if head1:
            cur.next = head1
        if head2:
            cur.next = head2
        return fake_head.next

    def cut_half(head):
        prev = slow = fast = head
    
        while fast and fast.next:
            prev = slow
            slow = slow.next
            fast = fast.next.next
        prev.next = None
        return head, slow

        
    if not head or not head.next:
        return head
    one, two = cut_half(head)
    one = merge_sort(one)
    two = merge_sort(two)
    return merge(one,two)      
```

```python
# given a singly linked list and a target value, remove all nodes in this list that contain this target
def remove_value(head, value):
    if not head:
        return head
    fake_head = ListNode(None)
    fake_head.next = head
    prev = fake_head
    curr = fake_head.next
    while curr:
        if curr.value == value:
            prev.next = curr.next
            curr.next = None
            curr = prev.next
        else:
            prev = curr
            curr = curr.next
    return fake_head.next
            
```
