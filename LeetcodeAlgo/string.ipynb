### Char Removal

Q1.1 Remove a/some particular chars from a string. eg. 'student',remove u and n -> 'stdet'

```python
def Remove(string, chars):
    new_string = []
    for i in range(len(string)):
        if string[i] not in chars:
            new_string.append(string[i]) 
            # 空间上比较浪费
    return ''.join(new_string)
        
```

```python
Remove('student', ['n', 'u'])
```

```python
# 两个指针实现
def Remove(string, chars):
    lst = list(string)
    i = j = 0
    while j < len(lst):
        if lst[j] not in chars:
            lst[i] = lst[j]
            i += 1
        j += 1

    return ''.join(lst[:i])
```

Q1.2 Remove all leading/trailing and duplicate empty spaces(only leave one empty space if duplicated spaces happen) from the input string. (must in place)

```python
def RemoveSpace(string):
    if not string or len(string) == 0:
        return string
    lst = list(string)
    i = j = 0
    while j < len(lst):
        if lst[j] != ' ':
            lst[i] = lst[j]
            i += 1
        else:
            if i != 0 and lst[i-1] != ' ':
                lst[i] = lst[j]
                i += 1
        j += 1
    if i > 0 and lst[i-1] == ' ':
        i -= 1
    return ''.join(lst[:i])
# 还可以优化
# 也可以新建list
```

```python
RemoveSpace('   abc  de  ')
```

### Char De-duplication

Q2.1 Remove duplicated and adjacent letters(leave only one letter in each duplicated section) in a string

eg: 'aabbazw' -> 'abazw'

```python
def remove_dup(string):
    if not string or len(string) < 2:
        return string
    lst = []
    j = 0
    while j < len(string):
        if j == 0:
            lst.append(string[j])
        else:
            if string[j-1] != string[j]:
                lst.append(string[j])
        j += 1
    return ''.join(lst)
```

```python
remove_dup('aabbazw')
```

```python
def remove_dup(string):
    if not string or len(string) < 2:
        return string
    lst = list(string)
    i = j = 0
    while j < len(string):
        if j == 0:
            lst[i] = lst[j]
            i += 1
        else:
            if lst[j-1] != lst[j]:
                lst[i] = lst[j]
                i += 1
        j += 1
    return ''.join(lst[:i])
```

Q2.1变种 Given a sorted array, remove duplicated and adjacent letters(leave only two letters in each duplicated section) in a string

eg. aaabbbccc -> aabbcc

```python
def remove_dup2(string):
    if not string or len(string) == 0:
        return string
    lst = list(string)
    i = j = 0
    while j < len(lst):
        if j == 0 or j == 1:
            lst[i] = lst[j]
            i += 1
        else:
            if lst[j] != lst[i-2]:
                lst[i] = lst[j]
                i += 1
        j += 1
    return ''.join(lst[:i]) 
```

```python
remove_dup2('aaabbbccc')
```
